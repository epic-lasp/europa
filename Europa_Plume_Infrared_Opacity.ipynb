## Direct plume detection by E-THEMIS

*Paul O. Hayne*

We estimate the thermal infrared opacity and radiance of a model Europa plume. Two components may contribute to this signal:

1. Ice particles
2. Water vapor

By analogy with Enceladus, solid particles may be small, $\sim 1~\mu$m (e.g., *Hansen et al.*, 2008). In this case, their infrared opacity is expected to be low, so they must be present in large quantities for detection. Water vapor is opaque at wavelengths $>15~\mu$m, but again emission by the cold, rarefied gas might be weak. Below we estimate these contributions to plume thermal emission, and evaluate detectability by E-THEMIS. Observations of the plume may be done in emission, with a cold background (e.g., space), or in absorption using a warm background (e.g., Jupiter). The analysis below focuses on emission, but the opacity numbers can be used to calculate plume absorption as well.

### Ice plume

For a plume with horizontal dimension $H$ and ice particle number density $n$ (m$^{-3}$), the line-of-sight column contains $N = nH$ particles per square meter. Given the single-scattering albedo $\varpi_0$ and extinction efficiency $Q_\mathrm{ext}$, the absorption efficiency is $Q_{\mathrm{abs}} = (1-\varpi_0)Q_{\mathrm{ext}}$. The optical depth for emission at wavelength $\lambda$ is

\begin{equation}
    \tau(\lambda) = \pi r^2 Q_\mathrm{abs}(\lambda)N
\end{equation}

where $r$ is the particle radius. As seen in emission, the infrared spectral radiance of the ice plume is then

\begin{equation}
    I_\lambda \approx \left(1-e^{-\tau}\right) B_\lambda(T)
\end{equation}

where $B_\lambda$ is the Planck function, and

\begin{equation}
    T = \left( \frac{(1-A)F}{4\epsilon \sigma} \right)^{1/4}
\end{equation}

is the equilibrium temperature for the solar flux $F = S/R_\mathrm{AU}^2$, given the solar distance $R_\mathrm{AU}$ and the solar constant, $S$. For albedo $A = 0.6$, $T \approx 100$ K. Each spectral channel has a different response function $f(\lambda)$, though here we use average values over the complete bandpass. We assume spherical particles and calculate scattering parameters from Mie theory, using optical constants for water ice from *Warren and Brandt* (2008).

```python
# Load H2O ice spectral data
# Based on optical constants from Warren and Brandt (2008) and Mie theory
wl = (np.genfromtxt(datadir+'/h2o_warren_LT1mm.wnr'))[:,0] # Wavelength [micron]
mie = (np.genfromtxt(datadir+'/h2o_warren_a3g4_00001p00.mie')) # Mie parameters
Q_ext = mie[:,0] # Extinction efficiency
w0 = mie[:,1] # Single-scattering albedo

# Calculate single-particle emissivity
e_particle = (1-w0)*Q_ext

# Equilibrium temperature of particles
T = europa.Teq()
```

```python
# Plot extinction efficiency over relevant wavelength range
fig1 = plt.figure(figsize=[10,5])

c1 = [0.8,0.2,0.2] # first color
c2 = [0.2,0.2,0.8] # second color
c3 = [0.2,0.6,0.4] # third color
wlmin, wlmax = 5.0, 100.0 # Approx. wavelength range covered by E-THEMIS

# Extinction plot
sub1 = fig1.add_subplot(121)
sub1.loglog(wl, Q_ext, color=c1, label='$Q_\mathrm{ext}$')
sub1.set_xlim([wlmin,wlmax])
sub1.set_ylim([0.5e-2,1])
sub1.set_xlabel('Wavelength ($\mu$m)')
sub1.set_ylabel('Extinction')

# Plot absorption efficiency
sub1.loglog(wl, e_particle, '--', color=c2, label='$Q_\mathrm{abs}$')

# E-THEMIS wavelength bands
wl_B1 = [7,14]
wl_B2 = [14,28]
wl_B3 = [28,70]
sub1.annotate(s='', xy=(wl_B1[0],1e-2), xytext=(wl_B1[1],1e-2), arrowprops=dict(arrowstyle='<->',fc='k'))
sub1.annotate(s='', xy=(wl_B2[0],1e-2), xytext=(wl_B2[1],1e-2), arrowprops=dict(arrowstyle='<->',fc='k'))
sub1.annotate(s='', xy=(wl_B3[0],1e-2), xytext=(wl_B3[1],1e-2), arrowprops=dict(arrowstyle='<->',fc='k'))
sub1.text(9, 1.2e-2, 'B1', fontsize=14)
sub1.text(18, 1.2e-2, 'B2', fontsize=14)
sub1.text(40, 1.2e-2, 'B3', fontsize=14)
sub1.legend(frameon=False)

# Single-scattering albedo
sub2 = fig1.add_subplot(122)
sub2.loglog(wl, w0, color=c1, label='$w_0$')
sub2.set_xlim([wlmin,wlmax])
sub2.set_ylim([0.5e-4,1])
sub2.set_xlabel('Wavelength ($\mu$m)')
sub2.set_ylabel('Single-scattering Albedo')

# E-THEMIS wavelength bands
sub2.annotate(s='', xy=(wl_B1[0],2e-4), xytext=(wl_B1[1],2e-4), arrowprops=dict(arrowstyle='<->',fc='k'))
sub2.annotate(s='', xy=(wl_B2[0],2e-4), xytext=(wl_B2[1],2e-4), arrowprops=dict(arrowstyle='<->',fc='k'))
sub2.annotate(s='', xy=(wl_B3[0],2e-4), xytext=(wl_B3[1],2e-4), arrowprops=dict(arrowstyle='<->',fc='k'))
sub2.text(9, 3e-4, 'B1', fontsize=14)
sub2.text(18, 3e-4, 'B2', fontsize=14)
sub2.text(40, 3e-4, 'B3', fontsize=14)
sub2.yaxis.tick_right()
sub2.yaxis.set_label_position('right')

plt.tight_layout()
plt.show()
```

**Figure (above):** Modeled extinction and scattering parameters for H$_2$O ice, using Mie theory and optical constants from *Warren and Brandt* (2008). E-THEMIS bandpass filters are indicated by horizontal arrows.

```python
# Band-averaged parameters
print('Band-averaged parameters:')
print('')
print('         B1      B2      B3')
print('       ---------------------')

# Extinction efficiency:
Q_ext_1 = np.mean(Q_ext[np.logical_and(wl > wl_B1[0], wl < wl_B1[1])])
Q_ext_2 = np.mean(Q_ext[np.logical_and(wl > wl_B2[0], wl < wl_B2[1])])
Q_ext_3 = np.mean(Q_ext[np.logical_and(wl > wl_B3[0], wl < wl_B3[1])])
print('Q_ext: {:.3f}   {:.3f}   {:.3f}'.format(Q_ext_1, Q_ext_2, Q_ext_3))

# Single-scattering albedo:
w0_1 = np.mean(w0[np.logical_and(wl > wl_B1[0], wl < wl_B1[1])])
w0_2 = np.mean(w0[np.logical_and(wl > wl_B2[0], wl < wl_B2[1])])
w0_3 = np.mean(w0[np.logical_and(wl > wl_B3[0], wl < wl_B3[1])])
print('w_0:   {:.3f}   {:.3f}   {:.3f}'.format(w0_1, w0_2, w0_3))

# Absorption efficiency:
Q_abs_1 = (1-w0_1)*Q_ext_1
Q_abs_2 = (1-w0_2)*Q_ext_2
Q_abs_3 = (1-w0_3)*Q_ext_3
```

---

We determine the signal-to-noise ratio ($\mathrm{SNR}$) on the plume measurement from the $\mathrm{SNR}$ at $T=100$ K ratio of the plume radiance in each spectral band, to that of a blackbody at the same temperature:

\begin{equation}
    \left( \mathrm{SNR} \right)_{\mathrm{ice},\lambda} = \frac{I_\lambda}{B_\lambda(T)}\left(\mathrm{SNR}\right)_{\mathrm{100~K},\lambda} \approx \left(1-e^{-\tau}\right)\left(\mathrm{SNR}\right)_{\mathrm{100~K},\lambda}
\end{equation}

```python
# Plume radiance calculation

NETD = np.array([10.0, 0.4, 0.2]) # Noise-equivalent temperature difference [K]
SNR = T/NETD
print('                      B1     B2     B3   ')
print('                    --------------------')
print('Assumed NETD @100 K: {:.2f}   {:.2f}   {:.2f}'.format(NETD[0],NETD[1],NETD[2]))

N = np.logspace(8,13,100) # column density [m-2]
r = 1.0e-6 # particle radius [m]

# Opacity
tau1 = np.pi*r**2*Q_abs_1*N
tau2 = np.pi*r**2*Q_abs_2*N
tau3 = np.pi*r**2*Q_abs_3*N

# Emissivity factor
f1 = (1-np.exp(-tau1))
f2 = (1-np.exp(-tau2))
f3 = (1-np.exp(-tau3))

# Plot results
fig2 = plt.figure(figsize=[5,5])
plt.loglog(N, f1*SNR[0], color=c1, label='B1')
plt.loglog(N, f2*SNR[1], color=c2, label='B2')
plt.loglog(N, f3*SNR[2], color=c3, label='B3')
plt.loglog(N, N*0+1, '--k', linewidth=0.5)
plt.xlabel('$N_{\mathrm{ice}}$ (m$^{-2}$)')
plt.ylabel(r'$\left(\mathrm{SNR}\right)_{\mathrm{ice},\lambda}$')
plt.ylim([1e-2,1e2])
plt.xlim([N.min(),N.max()])
plt.legend(frameon=False)
plt.show()
```

**Figure (above):** Signal-to-noise ratio (SNR) for plume thermal emission by the ice component, over a range of column mass. Each E-THEMIS spectral band is plotted separately.

The formula above also provides a critical threshold for the column density of ice particles $N_{\mathrm{crit}}$, where $\left(SNR\right)_{\mathrm{ice},\lambda} = 1$. Based on the parameters above, we find $N_{\mathrm{crit}} \sim 5\times10^{9}-1\times10^{11}$ m$^{-2}$. Given a total plume column mass $M$ (kg m$^{-2}$) and mass fraction of ice $x_\mathrm{ice} = M_{\mathrm{ice}}/M$, the number density is $N = x_\mathrm{ice} M/{\mu}$, where $\mu = \frac{4}{3}\pi r^3 \rho$ is the mass of one ice grain with density $\rho$.

If we have a measurement of the vapor, $N_{\mathrm{vapor}}$, the ice component is $N = N_{\mathrm{vapor}}\frac{\mu_\mathrm{H2O}}{\mu}x/\left(1 - x\right)$, where $\mu_\mathrm{H2O}$ is the mass of one water molecule. Note that $\mu_\mathrm{H2O}/\mu \sim 10^{-12}$.

As an example, we assume the *Roth et al.* (2014) value of $N_{\mathrm{vapor}} \sim 10^{20}$ m$^{-2}$ at an altitude of 200 km. We also assume $x = 0.5$, i.e. equal vapor and ice masses.

```python
# Example SNR calculation for Roth et al. (2014) plume
N_vap = 1e20 # vapor column density [m-2]
rho = 970.0 # ice density [kg.m-3]
m_h2o = pt.m_h2o # mass of one water molecule [kg]
m_ice = (4/3)*np.pi*r**3*rho # mass of one ice grain [kg]
x = 0.5 # ratio of ice/vapor

N = N_vap*(m_h2o/m_ice)*x/(1-x)
tau = np.pi*r**2*np.array([Q_abs_1,Q_abs_2,Q_abs_3])*N

print('Ice particle column at 200 km altitude: {:.2g} /m^2'.format(N))
print('')
print('                  B1        B2        B3')
print('               ----------------------------')
print('Optical depth:  {:.1e}   {:.1e}   {:.1e}'.format(tau[0],tau[1],tau[2]))
```

This is clearly much below detection. Closer to the surface, the column density may increase by a factor $\sim e^{(z-z_0)/H}$, where $H$ is the plume density scale height. *Roth et al.* assume $H = 200$ km, such that the column is $\sim 2.7$ times higher at the surface. In this case, the particle column density would be $N(0) \sim 10^8$ m$^{-2}$.

#### Conclusion: Ice Plume

Given the estimated vapor column density of $N_\mathrm{vap} \sim 10^{20}$ m$^{-2}$ at 200 km from *Roth et al.* (2014), it is unlikely that E-THEMIS would detect thermal emission from 1-$\mu$m ice particles, unless their total mass is $> 10$ times the vapor mass. Equivalently, if the plume source is $>10$ times the *Roth et al.* estimate, then E-THEMIS might be capable of detecting the ice particle emission in Band 3. Sub-micron particles are expected to be more transparent, and larger particles are less likely based on thermodynamic arguments (*Schmidt et al.*, 2008).

```python
# Show simulated plume, scaled using above factors
plume_data = (pd.read_table(datadir+'/europa_hotspot_plume_example_15yr.log10N.txt',delimiter=' ')).as_matrix()
plume_data = np.nan_to_num(plume_data)-3 # scale this to Roth et al. (2014)
xx = np.genfromtxt(datadir+'/europa_hotspot_plume_example_15yr.xR.txt')
yy = np.genfromtxt(datadir+'/europa_hotspot_plume_example_15yr.yR.txt')
N_vap = 10**plume_data

# Convert to ice particle column density
N = N_vap*(m_h2o/m_ice)*x/(1-x)
figN = plt.figure(figsize=[10,4.1])
pltN = figN.add_subplot(111)
with np.errstate(divide='ignore'):
    Nm = pltN.pcolormesh(xx, yy, np.log10(N), cmap='magma', vmin=-1, vmax=8)
cbN = figN.colorbar(Nm)
cbN.set_label(r'$\log_{10}N_\mathrm{ice}$ (m$^{-2}$)')
pltN.axis('equal')
pltN.set_xlim([-2,2])
pltN.set_ylim([0,2])
pltN.text(0.5, 1.7, 'Ice Column Density', color='w')
pltN.set_xlabel('$x/R$')
pltN.set_ylabel('$y/R$')
```

**Figure (above):** Ice particle column density for modeled plume based on vapor density reported by *Roth et al.* (2014), scaled to relative mass of ice and vapor.

---

```python
# Plot SNR for each band
# Convert to optical depth
tau1 = np.pi*r**2*Q_abs_1*N
tau2 = np.pi*r**2*Q_abs_2*N
tau3 = np.pi*r**2*Q_abs_3*N
SNR_p_1 = (1-np.exp(-tau1))*SNR[0]
SNR_p_2 = (1-np.exp(-tau2))*SNR[1]
SNR_p_3 = (1-np.exp(-tau3))*SNR[2]

# Set up figure
figp = plt.figure(figsize=[8,13.5])
psub1 = figp.add_subplot(311)
psub2 = figp.add_subplot(312)
psub3 = figp.add_subplot(313)
with np.errstate(divide='ignore'):
    pm1 = psub1.pcolormesh(xx, yy, np.log10(SNR_p_1), cmap='magma', vmin=-2, vmax=0)
    pm2 = psub2.pcolormesh(xx, yy, np.log10(SNR_p_2), cmap='magma', vmin=-2, vmax=0)
    pm3 = psub3.pcolormesh(xx, yy, np.log10(SNR_p_3), cmap='magma', vmin=-2, vmax=0)
    
psub1.axis('equal')
psub1.set_xlim([-2,2])
psub1.set_ylim([0,2])
psub1.set_xlabel('$x/R$')
psub1.set_ylabel('$y/R$')
psub1.text(1.25, 1.7, 'Band 1', color='w')

psub2.axis('equal')
psub2.set_xlim([-2,2])
psub2.set_ylim([0,2])
psub2.set_xlabel('$x/R$')
psub2.set_ylabel('$y/R$')
psub2.text(1.25, 1.7, 'Band 2', color='w')

psub3.axis('equal')
psub3.set_xlim([-2,2])
psub3.set_ylim([0,2])
psub3.set_xlabel('$x/R$')
psub3.set_ylabel('$y/R$')
psub3.text(1.25, 1.7, 'Band 3', color='w')

figp.subplots_adjust(right=0.8, hspace=0.35)
cbar_ax = figp.add_axes([0.9, 0.2, 0.04, 0.6])
cbm = figp.colorbar(pm3, cax=cbar_ax)
cbm.set_label(r'$\log_{10}\mathrm{SNR}$')
cbm.set_ticks(np.arange(-2,0.1,0.5))
```

**Figure (above):** Signal-to-noise ratio (SNR) for plume thermal emission by ice particles for each E-THEMIS band. Note the logarithmic scale, i.e. the top of the color scale corresponds to SNR = 1.
